```python
def tokenize_function(examples):
    # Prompt template - consistent format
    prompts = [f"翻译成泰语：{src}\n泰语：" for src in examples["source"]]
    targets = examples["translation"]  # Make sure this matches your column name
    
    # Create full texts (prompt + target + eos)
    full_texts = [prompt + tgt + tokenizer.eos_token for prompt, tgt in zip(prompts, targets)]

    # Tokenize full input (prompt + target)
    tokenized = tokenizer(
        full_texts,
        truncation=True,
        max_length=1140,
        padding='max_length',
        return_tensors=None,  # Don't return tensors yet
    )

    # Tokenize prompts only to get their lengths
    prompt_tokenized = tokenizer(
        prompts,
        add_special_tokens=True,  # IMPORTANT: Must match full tokenization
        truncation=True,
        max_length=1140,
        padding=False,  # Don't pad prompts
    )

    # Prepare masked labels
    labels = []
    for i, (input_ids, prompt_ids) in enumerate(zip(tokenized["input_ids"], prompt_tokenized["input_ids"])):
        label = input_ids.copy()
        prompt_length = len(prompt_ids)
        
        # Mask the prompt part (set to -100 so it's ignored in loss)
        label[:prompt_length] = [-100] * prompt_length
        
        # Handle padding tokens - also mask them
        for j, token_id in enumerate(label):
            if token_id == tokenizer.pad_token_id:
                label[j] = -100
        
        labels.append(label)

    tokenized["labels"] = labels
    return tokenized

# Apply tokenization to your dataset
tokenized_datasets = datasets.map(
    tokenize_function,
    batched=True,
    remove_columns=datasets["train"].column_names,
    desc="Tokenizing dataset"
)
```

```python
def print_trainable_parameters(model):
    trainable = 0
    total = 0
    for param in model.parameters():
        total += param.numel()
        if param.requires_grad:
            trainable += param.numel()
    print(f"Trainable parameters: {trainable:,}")
    print(f"Total parameters:     {total:,}")
    print(f"Trainable %:          {100 * trainable / total:.2f}%")

print_trainable_parameters(model)
```

```python
def translate_text(chinese_text):
    prompt = f"翻译成泰语：{chinese_text}\n泰语："
    
    # Tokenize input
    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=1140,
        padding="max_length",
    ).to("cuda")
    
    # Move to device
    if torch.cuda.is_available():
        inputs = {k: v.to(device) for k, v in inputs.items()}
        modelx.cuda()
    
    input_length = inputs['input_ids'].shape[1]
    
    # Generate translation
    with torch.no_grad():
        generated = modelx.generate(
            **inputs,
            temperature=0.3,  # Lower for more consistent translation
            # top_p=0.8,
            num_beams=10,  # ← This parameter
            # repetition_penalty=1.1,
            # do_sample = True,
            max_new_tokens=2048,
            pad_token_id=tokenizer.pad_token_id or tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
            early_stopping=True
        )
    
    # Extract only the generated part (skip the input prompt)
    generated_text = tokenizer.decode(
        generated[0][input_length:], 
        skip_special_tokens=True
    ).strip()
    
    return generated_text
```

```python
def translate_batch(chinese_texts, batch_size=8):
    # Handle pandas Series or DataFrame column
    if hasattr(chinese_texts, 'tolist'):
        text_list = chinese_texts.tolist()
    else:
        text_list = list(chinese_texts)
    
    results = []
    for i in tqdm.tqdm(range(0, len(text_list), batch_size)):
        batch = text_list[i:i+batch_size]
        prompts = [f"翻译成泰语：{text}\n泰语：" for text in batch]
        
        inputs = tokenizer(
            prompts,
            return_tensors="pt",
            truncation=True,
            max_length=1140,
            padding=True,  # Pad within batch only
        ).to(device)
        
        with torch.no_grad():
            generated = modelx.generate(
                **inputs,
                temperature=0.3,  # Lower for more consistent translation
                # top_p=0.8,
                num_beams=10,  # ← This parameter
                # repetition_penalty=1.1,
                # do_sample = True,
                max_new_tokens=2048,
                pad_token_id=tokenizer.pad_token_id or tokenizer.eos_token_id,
                eos_token_id=tokenizer.eos_token_id,
                early_stopping=True
            )
        
        # Decode batch results
        input_length = inputs['input_ids'].shape[1]
        for j, gen_seq in enumerate(generated):
            decoded = tokenizer.decode(
                gen_seq[input_length:], 
                skip_special_tokens=True
            ).strip()
            results.append(decoded)
    return results

# Usage examples for pandas:
# df['thai_translation'] = translate_batch(df['chinese_text'])
# thai_texts = translate_batch(df['source'])  # if column is named 'source'

# Alternative: Direct pandas integration
def translate_pandas_column(df, source_col, target_col='translation', batch_size=8):
    """
    Translate a pandas column directly
    """
    translations = translate_batch(df[source_col], batch_size=batch_size)
    df[target_col] = translations
    print(df)
    return df

df = translate_pandas_column(data_test,'source')
```
