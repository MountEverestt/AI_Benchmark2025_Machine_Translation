```python
from pythainlp.tokenize import word_tokenize

def remove_chinese_with_thai_tokenizer(text):
    tokens = word_tokenize(text, engine="newmm")
    cleaned_tokens = [
        token for token in tokens 
        if not any('\u4e00' <= char <= '\u9fff' for char in token)
    ]
    return ''.join(cleaned_tokens)
    
syn = {
    'à¹‰à¹à¸²' : 'à¹‰à¸³',
    'à¹ˆà¹à¸²' : 'à¹ˆà¸³',
    'à¹Šà¹à¸²' : 'à¹Šà¸³',
    'à¹‹à¹à¸²' : 'à¹‹à¸³',
    'à¹à¸²' : 'à¸³'
}
```

```python
import os
import tqdm
folder_path = "./summit_MT"
merge_data = []
for file in tqdm.tqdm(os.listdir(folder_path)):
    if(file == ".ipynb_checkpoints"): continue
    files = folder_path + '/' + file
    data = []
    print(files)
    with open(files, 'r', encoding='utf-8') as file:
        f = file.read().split('\n')
        for sentence in f:
            for k,v in syn.items(): sentence = sentence.replace(k,v)
            data.append(remove_chinese_with_thai_tokenizer(sentence.strip()))
    merge_data.append(data)
merge_data
```

```python
import pandas as pd

ensemble_data = pd.DataFrame([], columns=["translation"])
for idx, datax in enumerate(merge_data):
    data = pd.DataFrame(datax, columns=["translation"+str(idx)])
    ensemble_data = pd.concat([ensemble_data,data],axis=1)

ensemble_data = ensemble_data.drop("translation", axis=1) 
```

```python
from pythainlp.util import normalize

def correct_thai_sentence(text):
    # Normalize text and tokenize
    inputs = tokenizer(normalize(text), return_tensors="pt", padding=True, truncation=True)
    
    # Move input tensors to the same device as the model
    inputs = {k: v.to(device) for k, v in inputs.items()}
    
    # Run model inference
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Get predicted tokens
    predicted_tokens = outputs.logits.argmax(-1)
    
    # Decode and return corrected text
    corrected_text = tokenizer.decode(predicted_tokens[0], skip_special_tokens=True)
    return corrected_text

# Example usage
sentence = "à¸›à¸£à¸°à¹‚à¸¢à¸„à¹„à¸—à¸¢à¸—à¸µà¹ˆà¸¡à¸µà¸„à¸³à¸œà¸´à¸”"
corrected = correct_thai_sentence(sentence)
print(corrected)
```
