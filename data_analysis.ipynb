```python
import re
import tqdm

non_chinese = set()
def extract_non_chinese_chars(sentence):
    # Regex to match non-Chinese characters
    pattern = re.compile(r'[^\u4e00-\u9fff\u3400-\u4dbf\uff00-\uffef]')  # Adjust ranges as needed
    non_chinese_chars = pattern.findall(sentence)
    return non_chinese_chars

mem_idx = []
keep_mem = []
# Example usage
for idx, data in tqdm.tqdm(data_val.iterrows(), total = len(data_val['source'])):
    focus = extract_non_chinese_chars(data['source'])
    if(len(focus)): 
        mem_idx.append(idx) 
        keep_mem.append(focus)
    for c in focus: non_chinese.add(c)
print("Non-Chinese characters:", non_chinese)
```

```python
import numpy as np
from numba import cuda

# Thai Unicode ranges (0x0E00-0x0E7F, 0x0E80-0x0EFF)
THAI_RANGES = [(0x0E00, 0x0E7F), (0x0E80, 0x0EFF)]

@cuda.jit
def extract_non_thai_kernel(sentence, output_flags):
    i = cuda.grid(1)  # Thread index
    if i < len(sentence):
        char_code = ord(sentence[i])
        is_thai = False
        for (start, end) in THAI_RANGES:
            if start <= char_code < end:
                is_thai = True
                break
        output_flags[i] = 0 if is_thai else 1  # 1 = non-Thai

def extract_non_thai_gpu(sentence):
    # Convert sentence to a Unicode code-point array
    chars = np.array([ord(c) for c in sentence], dtype=np.int32)
    output_flags = np.zeros(len(sentence), dtype=np.int32)
    
    # GPU setup
    threads_per_block = 256
    blocks_per_grid = (len(sentence) + threads_per_block - 1) // threads_per_block
    
    # Run CUDA kernel
    extract_non_thai_kernel[blocks_per_grid, threads_per_block](sentence, output_flags)
    
    # Extract non-Thai characters
    non_thai_chars = [sentence[i] for i in range(len(sentence)) if output_flags[i] == 1]
    return ''.join(non_thai_chars)

# Example usage
sentence = "Hello ‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ ‰Ω†Â•Ω 123!"
non_thai = extract_non_thai_gpu(sentence)
print(non_thai)  # Output: "Hello ‰Ω†Â•Ω 123!"
```
